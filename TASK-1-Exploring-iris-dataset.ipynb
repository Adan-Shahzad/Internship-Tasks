```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
df = sns.load_dataset('iris')
```

```python
df.describe()
df.isnull().sum()
df.info()
```

```python
print(df.shape)
df.head()
df.columns.tolist()
```

```python
plt.figure(figsize=(8, 5))
sns.scatterplot(data=df, x='sepal_length', y='petal_length', hue='species', palette='Set1')
plt.title('Scatter Plot: Sepal Length vs Petal Length')
plt.tight_layout()
plt.show()
```

```python
df.drop(columns='species').hist(figsize=(10, 6), bins=20, color='steelblue', edgecolor='black')
plt.suptitle('Feature Distributions', fontsize=14, fontweight='bold')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
df.melt(id_vars='species', var_name='Feature', value_name='Value').pipe(
    lambda d: sns.boxplot(data=d, x='Feature', y='Value', hue='species', palette='Set2')
)
plt.title('Box Plot: Feature Distribution by Species')
plt.tight_layout()
plt.show()
```

```python
sns.pairplot(df, hue='species', palette='Set1')
plt.suptitle('Pair Plot - Iris Dataset', y=1.02, fontsize=14, fontweight='bold')
plt.show()
```

```python
plt.figure(figsize=(7, 5))
sns.heatmap(df.drop(columns='species').corr(), annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Heatmap')
plt.tight_layout()
plt.show()
```
